### Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Assuming `ndf` is your DataFrame
# Group by episode_number and calculate max & min for each column
gr = ndf.groupby('episode_number')[['ask_amount', 'ask_equity', 'ask_valuation']].agg(['max', 'min']).reset_index()

# Create 7 equal buckets for episodes 1-35
ndf['bucket'] = pd.cut(ndf['episode_number'], bins=[0, 5, 10, 15, 20, 25, 30, 35], 
                       labels=['1-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35'])

# Aggregate min and max values for each bucket
bucket_data = ndf.groupby('bucket')[['ask_amount', 'ask_equity', 'ask_valuation']].agg(['max', 'min']).reset_index()

# Set the figure size for better clarity
plt.figure(figsize=(14, 8))

# Creating a color palette for better visual distinction
colors = sns.color_palette("Set1", 3)

# Prepare positions for each bucket
bucket_positions = np.arange(len(bucket_data))  # position of buckets on the x-axis
bar_width = 0.2  # width of the bars

# Plot the bars for ask_amount, ask_equity, and ask_valuation

# Ask Amount Max and Min
plt.bar(bucket_positions - bar_width, bucket_data[('ask_amount', 'max')], width=bar_width, label='Ask Amount Max', color=colors[0])
plt.bar(bucket_positions, bucket_data[('ask_amount', 'min')], width=bar_width, label='Ask Amount Min', color=colors[1])

# Ask Equity Max and Min
plt.bar(bucket_positions + bar_width, bucket_data[('ask_equity', 'max')], width=bar_width, label='Ask Equity Max', color=colors[2])
plt.bar(bucket_positions + 2 * bar_width, bucket_data[('ask_equity', 'min')], width=bar_width, label='Ask Equity Min', color=colors[0])

# Ask Valuation Max and Min
plt.bar(bucket_positions + 3 * bar_width, bucket_data[('ask_valuation', 'max')], width=bar_width, label='Ask Valuation Max', color=colors[1])
plt.bar(bucket_positions + 4 * bar_width, bucket_data[('ask_valuation', 'min')], width=bar_width, label='Ask Valuation Min', color=colors[2])

# Labels and titles
plt.xlabel("Episode Ranges", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.title("Min and Max Values of Ask Amount, Ask Equity, and Ask Valuation per Episode Range", fontsize=14)

# Set x-ticks to correspond to the bucket positions and add labels
plt.xticks(bucket_positions + 2 * bar_width, bucket_data['bucket'])

# Add a legend
plt.legend()

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Q9. How many sharks participated in this show and What were their names? 


```python
shark_columns = [col for col in df.columns if '_present' in col]
shark_names = [col.replace('_present', '') for col in shark_columns]
participated = len(shark_names)
print("Number of sharks participated:", participated)
print("Names of sharks:", shark_names)
```

### Q10. Find appearance of each sharks? 


```python
appear = {shark: df[shark + '_present'].sum() for shark in shark_names}
print("Appearance of each shark:",appear)
```

```python
sorted_appear = {k: v for k, v in sorted(appear.items(), key=lambda item: item[1], reverse=True)}

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_appear.keys(), sorted_appear.values(), color='skyblue')

# Add count value labels on top of the bars
for i, v in enumerate(sorted_appear.values()):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom')

# Add labels and title
plt.xlabel('Sharks')
plt.ylabel('Appearance Count')
plt.title('Appearance of Each Shark Sorted by Count')

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Example data (replace with your actual 'invest' dictionary)
invest = {'ashneer': 21, 'anupam': 24, 'aman': 28, 'namita': 22, 'vineeta': 15, 'peyush': 27, 'ghazal': 7}
# Data for the donut chart
labels = invest.keys()
sizes = invest.values()

# Plotting the donut chart
fig, ax = plt.subplots(figsize=(5, 5))

# Create a pie chart and modify it into a donut chart by setting the 'wedgeprops' to have a 'width'
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')


plt.title('Investment Count for Each Shark')
plt.show()
```
