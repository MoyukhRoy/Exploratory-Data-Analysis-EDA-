### Q7. Brand names where deal equity is between 0 to 50?

### Q8. Find the number of brands participated in each episode? 

```python
b = df.groupby('episode_number')['brand_name'].count()
print(b.unique())
```

### Q9. How many sharks participated in this show and What were their names? 


```python
shark_columns = [col for col in df.columns if '_present' in col]
shark_names = [col.replace('_present', '') for col in shark_columns]
participated = len(shark_names)
print("Number of sharks participated:", participated)
print("Names of sharks:", shark_names)
```

# OR

```python
shark_columns = [col for col in df.columns if '_deal' in col]
shark_names = [col.replace('_deal', '') for col in shark_columns]
participated = len(shark_names)
print("Number of sharks participated:",participated)
print("Names of sharks:", shark_names)
```

### Q10. Find appearance of each sharks? 


```python
appear = {shark: df[shark + '_present'].sum() for shark in shark_names}
print("Appearance of each shark:",appear)
```

### Q11. How many entrepreneurs were present? 
 

```python
n = df['pitch_number'].nunique()
print("Total number of entrepreneurs:", n)
```

### Q12. How many times each shark invested the deal? 


```python
invest = {shark: df[shark + '_deal'].sum() for shark in shark_names}
print("Investment count for each shark:",invest)
```

### Q13. Find the equity percent that each sharks gets? 


```python
p = {
    shark: (df[df[shark + '_deal'] == 1]['equity_per_shark']).sum()
    for shark in shark_names
}
p
```

### Q14. Find the total number of amount invested in this show? 

```python
t = df['deal_amount'].sum()
print("Total amount invested in the show:", t.round())
```
